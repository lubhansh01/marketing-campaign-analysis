```python
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/lubhanshsharma/Downloads/marketing_campaign_data.csv")
```

```python
# High Income
df["High_Income"] = np.where(df["Income"] > 75000, 1, 0)

# Young Customer
df["Young_Customer"] = np.where(df["Age"] < 30, 1, 0)

# Campaign Responder
df["Campaign_Responder"] = df["Response"]

# High Web Engagement
df["High_Web_Engagement"] = np.where(df["NumWebVisitsMonth"] > 5, 1, 0)

# Family Customer
df["Family_Customer"] = np.where(df["Children"] > 0, 1, 0)

# High Spender (Top 10%)
threshold = df["Total_Spend"].quantile(0.90)
df["High_Spender"] = np.where(df["Total_Spend"] >= threshold, 1, 0)

df.head()
```

```python
segment_summary = df.groupby("Campaign_Responder").agg({
    "Income":"mean",
    "Total_Spend":"mean",
    "High_Income":"mean",
    "High_Spender":"mean"
})

segment_summary
```

```python
df.to_csv("/Users/lubhanshsharma/Downloads/marketing_campaign_data.csv", index=False)
print("✅ Final segmented data saved")
```
