```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/lubhanshsharma/Downloads/marketing_campaign_data.csv")
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df["Income"], kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Response", y="Income", data=df)
plt.title("Campaign Response vs Income")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Response", y="Total_Spend", data=df)
plt.title("Total Spend vs Campaign Response")
plt.show()
```

```python
channel_cols = [
    "NumWebPurchases","NumCatalogPurchases",
    "NumStorePurchases","NumDealsPurchases"
]

df[channel_cols].mean().plot(kind="bar", figsize=(7,4), title="Avg Channel Usage")
plt.show()
```
