```python
import pandas as pd
import numpy as np
from datetime import datetime

pd.set_option("display.max_columns", None)

# Load datasets
df = pd.read_csv("/Users/lubhanshsharma/Downloads/marketing_campaign_data.csv")
data_dict = pd.read_csv("/Users/lubhanshsharma/Downloads/marketing_data_dictionary.csv")

df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# Remove rows with missing Income (business critical field)
df = df[df["Income"].notna()]

# Remove unrealistic ages later (after age creation)
```

```python
CURRENT_YEAR = datetime.now().year

# Convert date
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], errors="coerce")

# Age
df["Age"] = CURRENT_YEAR - df["Year_Birth"]

# Remove unrealistic ages
df = df[(df["Age"] >= 18) & (df["Age"] <= 90)]

# Children
df["Children"] = df["Kidhome"] + df["Teenhome"]

# Total Spend
spend_cols = [
    "MntWines","MntFruits","MntMeatProducts",
    "MntFishProducts","MntSweetProducts","MntGoldProds"
]
df["Total_Spend"] = df[spend_cols].sum(axis=1)

# Total Purchases
purchase_cols = [
    "NumWebPurchases","NumCatalogPurchases",
    "NumStorePurchases","NumDealsPurchases"
]
df["Total_Purchases"] = df[purchase_cols].sum(axis=1)

# Customer Tenure (in days)
df["Customer_Tenure_Days"] = (datetime.now() - df["Dt_Customer"]).dt.days

df.head()
```

```python
df.to_csv("/Users/lubhanshsharma/Downloads/marketing_campaign_data.csv", index=False)
print("✅ Cleaned data saved successfully")
```
